```python
#1 copy the user and roles dataframes from the text
# users dataframe
users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users
```

```python
# roles dataframe
roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
#2 do a right join on the users and roles dataframes
users.merge(roles, how = 'right', left_on = 'role_id', right_on = 'id', indicator = True)
```

```python
#3 what is the result of doing an outer join
users.merge(roles, how = 'outer', left_on = 'role_id', right_on = 'id', indicator = True)
```

```python
#4 what happens if you drop the foreign keys and try to join
users_d_fk = users.drop(columns = 'role_id')
# joining using id in both dataframes, even though they are not the same thing
users.merge(roles, how = 'outer', left_on = 'id', right_on = 'id', indicator = True)
```
