```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#1 use matplotlib to plot the following equation and annotate the 0,0 origin:
# y = x^2 -x + 2
# generate x and y
x = range(-100, 100)
y = [((x ** 2) -x + 2) for x in x]
# plot x and y
plt.plot(x, y)
# add title and axis labels
plt.title('y = x^2 - x + 2')
plt.xlabel('x')
plt.ylabel('y')
# anotate the origin
plt.annotate('origin', xy = (0,0), xytext = (0, 3000), arrowprops = {})
```

```python
#2 create four separate charts for the following equations
#2-1 y = x ** 0.5
x = range(0, 100)
y = [(x ** 0.5) for x in x]
plt.plot(x, y)
plt.title('x ** 0.5')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#2-2
# y = x ** 3
x = range(0, 100)
y = [(x ** 3) for x in x]
plt.plot(x, y)
plt.title('y = x ** 3')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#2-3
# y = 2 ** x
x = range(0, 100)
y = [(2 ** x) for x in x]
plt.plot(x,y)
plt.title('y = 2 ** x')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#2-4
# y = 1 / (x + 1)
x = range(0, 100)
y = [(1 / (x + 1)) for x in x]
plt.plot(x, y)
plt.title('1 / (x + 1)')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#3 combine the four plots from question 2 as subplots
# organizing the series that I will be plotting
x = range(0, 100)
y1 = [(x ** 0.5) for x in x]
y2 = [(x ** 3) for x in x]
y3 = [(2 ** x) for x in x]
y4 = [(1 / (x + 1)) for x in x]

#setting up the rows and columns for the subplots
n_rows = 2
n_cols = 2

# the first subplot
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y1)
plt.title('x ** 0.5')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2)
plt.title('x ** 3')

# the third subplot
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3)
plt.title('2 ** x')

# the fourth subplot
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y4)
plt.title('1 / (x + 1')

plt.show()
```

```python
#4 recreate the figure from question 2,
#  but with different colors for the points,
#  a title for the overall figure,
#  and a legend for the figure

# organizing the series that I will be plotting
x = range(0, 100)
y1 = [(x ** 0.5) for x in x]
y2 = [(x ** 3) for x in x]
y3 = [(2 ** x) for x in x]
y4 = [(1 / (x + 1)) for x in x]

#setting up the rows and columns for the subplots
n_rows = 2
n_cols = 2

# the first subplot
plt.subplot(n_rows, n_cols, 1)
y_1,= plt.plot(x, y1, color = 'blue', label = 'Line 1')
plt.title('x ** 0.5')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
y_2,= plt.plot(x, y2, color = 'red', label = 'Line 2')
plt.title('x ** 3')

# the third subplot
plt.subplot(n_rows, n_cols, 3)
y_3,= plt.plot(x, y3, color = 'green', label = 'Line 3')
plt.title('2 ** x')

# the fourth subplot
plt.subplot(n_rows, n_cols, 4)
y_4,= plt.plot(x, y4, color = 'yellow', label = 'Line 4')
plt.title('1 / (x + 1')

# creating title for overall figure
plt.suptitle('four subplots')

# creating a legend for the figure
plt.legend(handles = [y_1, y_2, y_3, y_4])
# another way of getting a legend if it isn't possiple to set the label of the handle
#plt.legend([y_1, y_2, y_3, y_4], ['y_1', 'y_2', 'y_3', 'y_4'])

plt.show()

```
