```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#5 make a new chart
# title is 'Big O Notation'
# x axis is 'Elements'
# y axis is 'Operations'
# label curves or make a legend for curves
# use LaTex notation when possible

# practice graph to get table title and labels
x = range(0, 100)
y = range(0, 100)

plt.plot(x, y)
plt.title('$Big\;O\;Notation$')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')
```

```python
#5
# curves to graph
# y = 0 * n + 1     label as 'O(1)'
# y = log(n)        label as 'O(log n)'
# y = n             label as 'O(n)'
# y = n * log(n)    label as 'O(n log n)'
# y = n ** 2        label as 'O(n^2)'
# y = 2 ** n        label as 'O(2^n)'
# y = n!            label as 'O(n!)'
# y = n ** n.       label as 'O(n^n)'

x = range(1, 100)

y = [0 * n + 1 for n in x]
plt.plot(x, y, label = '$O(1)$')

y = [np.log(n) for n in x]
plt.plot(x, y, label = '$O(log\;n)')

y = [n for n in x]
plt.plot(x, y, label = '$O(n)$')

y = [n * np.log(n) for n in x]
plt.plot(x, y, label = '$O(n\;log\;n)')

y = [n ** 2 for n in x]
plt.plot(x, y, label = 'O(n^2)')

plt.title('$Big\;O\;Notation$')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')

plt.legend(loc = 'upper right')
#plt.legend(handles = [eq_1, eq_2])
```
